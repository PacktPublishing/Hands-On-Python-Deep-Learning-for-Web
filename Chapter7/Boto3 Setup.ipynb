## Loading up the necessary environment variables

```python
import os

aws_access_key_id= os.environ['aws_access_key_id']
aws_secret_access_key = os.environ['aws_secret_access_key']
```

## Calling `boto3` to interact with Amazon s3 with your credentials

```python
import boto3

s3 = boto3.resource(
    's3',
    aws_access_key_id=aws_access_key_id,
    aws_secret_access_key=aws_secret_access_key
)
```

## Enlisting the bucket names

```python
for bucket in s3.buckets.all():
    print(bucket.name)
```

## Uploading an image to your bucket

```python
data = open('my_image.jpeg', 'rb')
s3.Bucket('demo-bucket-sayak').put_object(Key='my_image.jpeg', Body=data)
```
